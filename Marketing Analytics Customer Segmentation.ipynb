## 4: Customer Segmentation

```python
from sklearn.cluster import KMeans

# Select relevant features for segmentation
X = ifood[['Income', 'Kidhome', 'Teenhome', 'Recency', 'MntWines', 'MntFruits', 'MntMeatProducts',
              'MntFishProducts', 'MntSweetProducts', 'MntGoldProds', 'NumDealsPurchases', 'NumWebPurchases',
              'NumCatalogPurchases', 'NumStorePurchases', 'NumWebVisitsMonth', 'AcceptedCmp3', 'AcceptedCmp4',
              'AcceptedCmp5', 'AcceptedCmp1', 'AcceptedCmp2', 'Complain', 'Z_CostContact', 'Z_Revenue',
              'Age', 'Customer_Days', 'marital_Divorced', 'marital_Married', 'marital_Single', 'marital_Together',
              'marital_Widow', 'education_2n Cycle', 'education_Basic', 'education_Graduation', 'education_Master',
              'education_PhD', 'MntTotal', 'MntRegularProds', 'AcceptedCmpOverall']]

# Standardize the features

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Determine optimal number of clusters using Elbow Method

from yellowbrick.cluster import KElbowVisualizer

model = KMeans()
visualizer = KElbowVisualizer(model, k=(2,10))
visualizer.fit(X_scaled)
visualizer.show()

# Fit K-means clustering model

kmeans = KMeans(n_clusters=3, random_state=42)  
kmeans.fit(X_scaled)

# Add cluster labels to the dataframe

ifood['cluster'] = kmeans.labels_
```

```python
# Calculate average purchase value for each cluster

avg_purchase_by_cluster = ifood.groupby('cluster')['MntTotal'].mean()

print("Average Purchase Value by Cluster:")
print(avg_purchase_by_cluster)
```

```python
# Select relevant features for segmentation

X = ifood[['MntTotal', 'NumStorePurchases', 'NumWebPurchases', 'NumCatalogPurchases']]  

# Standardize the features

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Determine optimal number of clusters using Elbow Method

model = KMeans()
visualizer = KElbowVisualizer(model, k=(2,10))
visualizer.fit(X_scaled)
visualizer.show()

# Fit K-means clustering model

kmeans = KMeans(n_clusters=3, random_state=42)  # Adjust number of clusters as needed
kmeans.fit(X_scaled)

# Add cluster labels to the dataframe

ifood['cluster'] = kmeans.labels_
```

## 5: Visualization

```python
# Scatter plot of Age vs. Income colored by Cluster

plt.figure(figsize=(10, 6))
plt.scatter(ifood['Age'], ifood['Income'], c=ifood['cluster'], cmap='viridis')
plt.xlabel('Age')
plt.ylabel('Income')
plt.title('Customer Segmentation based on Age and Income')
plt.colorbar(label='Cluster')
plt.show()
```

```python
#Bar chart of Mean Purchase Value by Cluster

segment_avg_purchase = ifood.groupby('cluster')['MntTotal'].mean()

plt.figure(figsize=(8, 5))
segment_avg_purchase.plot(kind='bar', color='skyblue')
plt.xlabel('Cluster')
plt.ylabel('Average Purchase Value')
plt.title('Average Purchase Value by Cluster')
plt.xticks(rotation=0)
plt.show()
```

```python
# Bar chart of Purchase Frequency by Cluster

purchase_frequency = ifood.groupby('cluster').size() / len(ifood)

plt.figure(figsize=(8, 5))
purchase_frequency.plot(kind='bar', color='salmon')
plt.xlabel('Cluster')
plt.ylabel('Purchase Frequency')
plt.title('Purchase Frequency by Cluster')
plt.xticks(rotation=0)
plt.show()
```

```python
# Bar chart of Response Rate by Cluster

response_rate = ifood.groupby('cluster')['Response'].mean()

plt.figure(figsize=(8, 5))
response_rate.plot(kind='bar', color='lightgreen')
plt.xlabel('Cluster')
plt.ylabel('Response Rate')
plt.title('Response Rate by Cluster')
plt.xticks(rotation=0)
plt.ylim(0, 1)
plt.show()
```

```python
# Bar chart of Teenhome distribution by cluster

sns.countplot(x='Teenhome', hue='cluster', data=ifood)
plt.title('Teenhome Distribution by Cluster')
plt.xlabel('Number of Teens at Home')
plt.ylabel('Count')
plt.show()
```

```python
# Bar chart of AcceptedCmpOverall distribution by cluster

sns.countplot(x='AcceptedCmpOverall', hue='cluster', data=ifood)
plt.title('Accepted Campaigns Distribution by Cluster')
plt.xlabel('Accepted Campaigns')
plt.ylabel('Count')
plt.show()
```
